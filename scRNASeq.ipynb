# Analysis Part

```python
import anndata
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as mpatches
import scanpy as sc
from scipy import stats
from matplotlib import cm
from matplotlib.lines import Line2D
```

```python
# Filtering Parameters as mentioned in the paper
cell_threshold_min = 500
cell_threshold_max = 7000
mito_criteria = 10
gene_threshold = 101
```

```python
actual_data = anndata.read_h5ad("../kallisto/final_Sample.h5ad")
actual_data.var['gene_id'] = actual_data.var.index.values

t2g = pd.read_csv("../kallisto/t2g.txt", header=None, names=['tid','gene_id','gene_name'], sep="\t")
t2g.index = t2g.gene_id
t2g = t2g.loc[~t2g.index.duplicated(keep='first')]

actual_data.var['gene_name'] = actual_data.var.gene_id.map(t2g["gene_name"])
actual_data.var.index = actual_data.var['gene_name']
actual_data.var_names_make_unique()
```

```python
actual_data.obs
```

```python
actual_data.var
```

## Quality Check

```python
def flatten(arr):
    return np.asarray(arr).reshape(-1)

actual_data.obs['cell_counts'] = actual_data.X.sum(axis=1)
actual_data.var['gene_counts'] = flatten(actual_data.X.sum(axis = 0))

actual_data.obs["n_genes"] = flatten((actual_data.X>0).sum(axis=1))
actual_data.var["n_cells"] = flatten((actual_data.X>0).sum(axis=0))
```

### Filtering Mitochondrial Genes

```python
mito_genes = actual_data.var_names.str.startswith('MT-')
actual_data.obs["percent_mito"] = actual_data[:,mito_genes].X.sum(axis=1)/actual_data.X.sum(axis=1)*100
```

### Knee Plot, Distribution of cell counts, mito gene percentage

```python
expected_num_cells = 7000
knee = np.sort(flatten(actual_data.X.sum(axis=1)))[::-1]

x = knee
y = range(len(knee))

fig, ax = plt.subplots(figsize=(5, 5))

ax.loglog(x, y, linewidth=4, color="red")
ax.axvline(x=knee[expected_num_cells], linewidth=3, color="k")
ax.axhline(y=expected_num_cells, linewidth=3, color="k")

ax.set_xlabel("UMI Counts")
ax.set_ylabel("Set of Barcodes")

plt.show()
```

```python
sc.pl.scatter(actual_data, x='cell_counts',y='percent_mito')
sc.pl.scatter(actual_data, x='cell_counts', y='n_genes')
```

```python
a = np.log10(actual_data.obs["cell_counts"])
result = plt.hist(a, bins='auto',edgecolor='k', alpha=0.65)
plt.axvline(np.log10(cell_threshold_min), color="k", linestyle="dashed")
plt.axvline(np.log10(cell_threshold_max), color="k", linestyle="dashed")
plt.xlabel("UMI Counts (log10 scale)")
plt.ylabel("Frequency")
plt.show()
```

```python
actual_data = actual_data[actual_data.obs.cell_counts < cell_threshold_max, :]
actual_data = actual_data[actual_data.obs.percent_mito < mito_criteria, :]
```
